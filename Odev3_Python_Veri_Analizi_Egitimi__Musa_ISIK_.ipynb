```python
import yfinance as yf
import plotly.express as px
import plotly.graph_objects as go
import pandas as pd
from plotly.tools import make_subplots
import warnings
warnings.filterwarnings("ignore")
```

### 2021 EUR/TRY kur tablosunun incelenmesi

1. <i> Yahoo Finance</i> sitesinden 2021 yılına ait EUR/TRY kur değerlerini indiriniz

```python
euro_try=yf.download('EURTRY=X',start='2021-01-01',end='2021-12-31')
euro_try
```

2. Veriyi özet metodları ile inceleyiniz

```python
euro_try.head()
```

```python
euro_try.info()
```

```python
euro_try.describe()
```

3. Veri temizleme: <br>
	• Tabloda sadece "Adj Close" (düzeltilmiş kapanış) sütununu bırakınız.<br>
	• "Adj Close" sütununu " Kapanış (TRY)" olarak yeniden adlandırınız.

```python
euro_try=euro_try[['Adj Close']]
euro_try.columns=['Kapanış (TRY)']
euro_try
```

4. Tabloya yeni sütun ekleme:<br>
	• Tabloya yeni bir sütun ekleyerek Kapanış (TRY) değerlerini 1 Ocak 2021 tarihindeki kapanış değerine göre normalize ediniz.<br>
	• BIST30 endeksinin 2021 yılı boyunca değişimini yorumlayınız.

```python
euro_try['Normalize EURO/TRY']=euro_try['Kapanış (TRY)']/euro_try.loc[euro_try.index[0],'Kapanış (TRY)']
euro_try
```

5. Veriyi grafik olarak inceleyiniz:<br>
	• Tabloyu px grafiği olarak görselleştiriniz.<br>
	• Tabloyu go. grafiği olarak görselleştiriniz.

```python
fig1_line=px.line(euro_try,x=euro_try.index,y='Normalize EURO/TRY')
fig2_bar=px.bar(euro_try,x=euro_try.index,y='Normalize EURO/TRY')
fig3_scatter=px.scatter(euro_try,x=euro_try.index,y='Normalize EURO/TRY')
fig1_line.show()
fig2_bar.show()
fig3_scatter.show()
```

```python
euro_try['degisim']=euro_try['Normalize EURO/TRY'].diff(1)
euro_try['d_yon']=euro_try['degisim'].map(lambda x:'dusen gun' if x<0 else 'yukselen gun')
euro_try['degisim%']=(euro_try['degisim']/euro_try['Normalize EURO/TRY'].shift(1))*100 

px.bar(euro_try,y='degisim%',
       color='d_yon',
       color_discrete_map={'dusen gun' : 'red','yukselen gun' : 'green'} )
```

```python
figure= go.Figure()
figure.add_trace(go.Bar(x=euro_try.index,y=euro_try['Normalize EURO/TRY'],name='Normalize EURO/TRY'))
```

### 2021 ve 2022 değerlerini birlikte inceleme

1. <i> Yahoo Finance</i> sitesinden 2022 BIST30 ve EUR/TRY kur değerlerini indiriniz

```python
dataset=['XU030.IS','EURTRY=X']
veri_2022=pd.DataFrame()
for i in dataset:
  data=yf.download(i,start='2022-01-01',end='2022-12-31')
  data=data['Adj Close'].rename(i)
  veri_2022=pd.concat([veri_2022,data],axis=1)

veri_2022=veri_2022[veri_2022['XU030.IS']>0]
veri_2022
```

2. Indirdiğiniz verileri 2021 verilerini temizleyip düzenlediğiniz üzere hazırlayınız.

```python
dataset_2021=['XU030.IS','EURTRY=X']
veri_2021=pd.DataFrame()

for i in dataset_2021:
  data=yf.download(i,start='2021-01-01',end='2021-12-31')
  data=data['Adj Close'].rename(i)
  veri_2021=pd.concat([veri_2021,data],axis=1)

veri_2021=veri_2021[veri_2021['XU030.IS']>0]
veri_2021
```

3. 2021 ve 2022 veri tablolarını tek bir tablo haline getiriniz.

```python
Veri=pd.concat([veri_2021,veri_2022],axis=0)
Veri
```

```python
Veri['Normalize BIST30'] = Veri['XU030.IS']/Veri.iloc[0,0]
Veri["Normalize EUR/TRY"] = Veri["EURTRY=X"]/ Veri.loc[Veri.index[0],"EURTRY=X"]
Veri["Degisim"] = Veri.Normalize.diff(1)
Veri["D_yon"] = Veri.Degisim.map(lambda x: "Düşen gün" if x<0 else "Yükselen gün")
Veri["Degisim%"] = (Veri.Degisim/Veri.Normalize)*100
Veri
```

4. Veriyi grafik olarak inceleyiniz:<br>
	• Tabloları px grafiği olarak görselleştiriniz.<br>
	• Tabloları go. grafiği olarak görselleştiriniz.<br>
	• Tabloları subplots grafiği olarak görselleştiriniz.

```python
Veri_line = px.line(Veri,y=Veri.Normalize,labels={'index':'Tarih',"Normalize" : "BIST 30 Endeksi (TRY)"})
Veri_line.show()
```

```python
fig=go.Figure()
fig.add_trace(go.Bar(x=Veri.index,y=Veri['Normalize'],name='Normalize BIST30'))
fig.add_trace(go.Scatter(x=Veri.index,y=Veri['Normalize EUR/TRY'],name='Normalize EUR/TRY'))
fig.update_layout(title='Normalize BIST 30 ve Normalize Kur',
                  xaxis_title='Tarih',xaxis_dtick='M1',
                  yaxis_title='Kapanış(TRY)')
```

```python
fig = make_subplots(rows=1, cols=2)

fig.add_trace(go.Scatter(x = Veri.index , y= Veri["Normalize"], name="Normalize"),
              row=1, col=1)

fig.add_trace(go.Bar(x = Veri.index , y= Veri["Normalize EUR/TRY"], name="Normalize EUR/TRY"),
              row=1, col=2)

fig.show()
```
